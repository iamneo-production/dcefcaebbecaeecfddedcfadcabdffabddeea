```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 

# Sample data (replace this with your actual patient vital signs data)
data = {
    'timestamp': ['2023-08-01 08:00', '2023-08-01 09:00', '2023-08-01 10:00', '2023-08-01 11:00'],
    'heart_rate': [75, 80, 85, 90],
    'blood_pressure': [120, 130, 125, 140],
    'temperature': [98.6, 98.7, 98.5, 99.0]
}

 

# Create a DataFrame from the sample data
df = pd.DataFrame(data)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)

 

# Set the style for Seaborn
sns.set(style="whitegrid")

 

# Create line plots for vital signs over time
plt.figure(figsize=(12, 6))

 

sns.lineplot(data=df['heart_rate'], label='Heart Rate')
sns.lineplot(data=df['blood_pressure'], label='Blood Pressure')
sns.lineplot(data=df['temperature'], label='Temperature')

 

plt.title('Patient Vital Signs Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
