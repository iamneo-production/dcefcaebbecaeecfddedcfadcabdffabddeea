```python
import numpy as np

import matplotlib.pyplot as plt

 

# Generate example data (replace this with your actual data)

np.random.seed(0)

num_samples = 100

num_parameters = 3

 

data = np.random.rand(num_samples, num_parameters) * 100  # Example data

 

# Identify abnormal time periods

abnormal_periods = []

for param_idx in range(num_parameters):

    param_data = data[:, param_idx]

    mean = np.mean(param_data)

    std = np.std(param_data)

    abnormal_indices = np.where(np.abs(param_data - mean) > 3 * std)[0]

    abnormal_periods.append(abnormal_indices)

 

# Create scatter plot

plt.figure(figsize=(10, 6))

 

for param_idx in range(num_parameters):

    plt.scatter(data[:, param_idx], param_idx * np.ones(num_samples), label=f'Parameter {param_idx + 1}')

 

# Highlight abnormal points

for param_idx, abnormal_indices in enumerate(abnormal_periods):

    plt.scatter(data[abnormal_indices, param_idx], param_idx * np.ones(len(abnormal_indices)),

                color='red', label=f'Abnormal Parameter {param_idx + 1}', marker='x')

 

plt.yticks(range(num_parameters), [f'Parameter {param_idx + 1}' for param_idx in range(num_parameters)])

plt.xlabel('Value')

plt.title('Scatter Plot of Parameters')

plt.legend()

plt.tight_layout()

plt.show()
```
